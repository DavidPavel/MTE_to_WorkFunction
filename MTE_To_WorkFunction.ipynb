## Calculating the work function from the MTE
We can calculate the work function from the MTE value, as mentioned in **"Cite as: Appl. Phys. Lett. 103, 103504 (2013); https://doi.org/10.1063/1.4820132"**

$MTE = \frac{1}{3} \left( h\nu - \phi \right)$

where $h\nu$ is the photon energy and $\phi$ is the work function. Then we can arrange:

$\phi = h \nu - \left(3\cdot MTE\right)$

```python
h = 4.135667696e-15    # eV·s
c = float(299792458)   # m/s

labda_1 = np.array([256.0,261.0,266.0,271.0,276.0,281.0,286.0,291.0,296.0,301.0,306.0,311.0])
labda_2 = np.array([256.0,261.0,266.0,271.0,276.0,281.0,286.0,291.0,296.0,301.0,306.0,311.0,316.0,321.0,326.0,331.0])
MTE_1 = np.array([268.8,230.2,199.0,168.0,156.6,133.5,124.4,106.1,87.1,67.2,55.2,36.7, 29.8,23.3,23.2,27.3])  # 17th OCT - Cu (111)
MTE_2 = np.array([266.1,243.9,222.9,202.1,174.2,157.9,137.2,116.6,99.7,83.5,64.4,48.8,37.0,27.9,25.5,23.7])    # 13th NOV - Cu (111)

# MTE_Cu100= np.array([50.7, 79.4, 94.5, 129.1, 178.7, 197.5, 205.2, 305.8, 270.1])   #  3rd March Cu (100)
                                                                        # We need to invert, due we take the me-
                                                                        # asurements the other way around 
MTE_Cu100= np.array([270.1, 305.8, 205.2, 197.5, 178.7, 129.1, 94.5, 79.4, 50.7])        
labda_Cu100 = np.array([226.0,231.0,236.0,241.0,246.0,251.0,256.0,261.0,266.0])




v = c/(labda_2*1e-9)     # Because we are working in nm 

print("Frecuency = ",v)
print("----------------------------------------")

phi = (h*v) - (3*MTE_2*0.001)   # Because we have the measurements in meV

print("Work Function values = ", phi)

#Cu(111)

fig = plt.figure(figsize = (6, 4))
#plt.errorbar(Frequency, visual_V0, d_visual_V0, fmt='.')
plt.plot(labda_2, phi, marker='.')#label="Line Fit",)
#plt.plot(Frequency, func(Frequency))
#plt.xlabel("Frequency [Hz]")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Work function [eV]")
plt.grid(color="gray")
```
